# LDA Topic Modeling by Month

```python
import pandas as pd
from ast import literal_eval
from gensim import corpora, models, utils
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Cleaned_Data.csv')
```

```python
df.cleaned_text = df.cleaned_text.apply(literal_eval)
```

```python
df.date = pd.to_datetime(df.date)
```

```python
df_sub = df[['date','content']]
```

```python
grouped = df.set_index('date').groupby(pd.Grouper(freq='M'))
```

```python
def get_month(index):
    return [grouped.get_group(x) for x in grouped.groups][index].cleaned_text
```

## LDA

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
def dummy(doc):
    return doc
```

```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
def pyLDA(month, num_topics=16):
    tf_vectorizer = CountVectorizer(max_df = 0.5, min_df = 5, tokenizer=dummy,preprocessor=dummy)
    m0 = get_month(month)
    dtm_tf = tf_vectorizer.fit_transform(m0)
    lda_tf = LatentDirichletAllocation(num_topics, batch_size = 200, evaluate_every=-1, learning_method='online', learning_offset = 10.0)
    lda_tf.fit(dtm_tf)
    return pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer, mds='mmds')
```
